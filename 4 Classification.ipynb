# Classification

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
datafile = "/content/sample_data/titanic.csv"
data = pd.read_csv(datafile)

data.head()
```

```python
data['Gender']=data['Sex'].apply(lambda a: 1 if a=='male' else 0)

data=data.drop(['PassengerId','Name','Cabin','Embarked','Ticket','Sex'],axis=1)

data=data.dropna()

data.head()
```

```python
# Normalize Data
scaler = MinMaxScaler(feature_range=(0, 1))

scaled_data = scaler.fit_transform(data)

scaled_df = pd.DataFrame(scaled_data, columns=data.columns.values)

# Create the X and y arrays
X = scaled_df.drop(['Survived'],axis=1).as_matrix()
y = scaled_df['Survived'].as_matrix()

# Split the data set in a training set (70%) and a test set (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

X_train, X_test, y_train, y_test 
```

```python
print(data.columns)
print(len(data.columns))
```

```python
import keras
from keras.layers import *
from keras.models import Sequential
```

```python
model=Sequential()
model.add(Dense(10,input_dim=6,activation='relu'))
model.add(Dense(100))
model.add(Dense(1))
```

```python
# Binary Classification Loss
model.compile(loss="binary_crossentropy",optimizer='adam')

# Batch Size: Size of data to be passed for each epoch
# Validation Data: Test Data for Analysis
model.fit(X_train,y_train,shuffle=True, batch_size=32, validation_data=(X_test,y_test),epochs=10)
```

```python
model.evaluate(X_test,y_test)
```

```python
model.predict(X_test)
```

```python
prediction=model.predict_classes(X_test).flatten()

pd.DataFrame({"Pred":prediction, "Actual":y_test},columns=['Pred',"Actual"])
```

## Additional Performance Metrics

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,prediction)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,prediction)
```
